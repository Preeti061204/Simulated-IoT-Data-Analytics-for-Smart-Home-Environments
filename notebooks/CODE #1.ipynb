```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
import warnings

warnings.filterwarnings('ignore')

# Load dataset
file_path = r"C:\Users\onush\OneDrive\Documents\IOT\Combined.xlsx"
df = pd.read_excel(file_path)
print("âœ… File loaded. Columns:", df.columns.tolist())

# Define columns
feature_cols = [
    'LivingRoom_Energy_kWh', 'Bathroom1_Energy_kWh', 'Bathroom2_Energy_kWh',
    'Bedroom1_Energy_kWh', 'Bedroom2_Energy_kWh', 'Kitchen_Energy_kWh'
]
target_col = 'Total_Energy_kWh'
datetime_col = 'Datetime'

df_clean = df.dropna(subset=feature_cols + [target_col])

# Correlation Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df_clean[feature_cols + [target_col]].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()

# Features and target
X = df_clean[feature_cols]
y = df_clean[target_col]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Models
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
    'Support Vector Regressor': SVR(),
    'KNN Regressor': KNeighborsRegressor(n_neighbors=5),
    'MLP Regressor': MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
}

results = {}

# Evaluation
print("\nðŸ” MODEL EVALUATION STARTS HERE")
for name, model in models.items():
    print(f"\n\n{'='*60}\nðŸš€ Starting Evaluation for: {name}\n{'='*60}")

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    accuracy = 100 - mape * 100
    residuals = y_test - y_pred

    results[name] = {
        'model': model,
        'y_pred': y_pred,
        'mse': mse,
        'r2': r2,
        'accuracy': accuracy,
        'residuals': residuals
    }

    print(f"ðŸ“Š MSE={mse:.2f}, RÂ²={r2:.2f}, Accuracy={accuracy:.2f}%")

    # Actual vs Predicted
    plt.figure(figsize=(6, 4))
    plt.scatter(y_test, y_pred, alpha=0.6)
    plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.title(f'{name} - Actual vs Predicted')
    plt.tight_layout()
    plt.show()

    # Feature Importance
    model_obj = model
    if hasattr(model_obj, 'coef_'):
        coef_df = pd.DataFrame({'Feature': feature_cols, 'Importance': model_obj.coef_})
    elif hasattr(model_obj, 'feature_importances_'):
        coef_df = pd.DataFrame({'Feature': feature_cols, 'Importance': model_obj.feature_importances_})
    else:
        coef_df = None

    if coef_df is not None:
        coef_df['Importance'] = coef_df['Importance'].astype(float)
        coef_df = coef_df.sort_values(by='Importance', key=abs, ascending=False)

        plt.figure(figsize=(6, 4))
        sns.barplot(x='Importance', y='Feature', data=coef_df)
        plt.title(f'{name} - Feature Importance')
        plt.tight_layout()
        plt.show()

        print(f"\n{name} - Feature Importances:")
        print(coef_df.round(4))

    # Residual Plot
    plt.figure(figsize=(6, 4))
    plt.scatter(y_pred, residuals, alpha=0.6)
    plt.axhline(0, color='red', linestyle='--')
    plt.xlabel('Predicted')
    plt.ylabel('Residuals')
    plt.title(f'{name} - Residual Plot')
    plt.tight_layout()
    plt.show()

    # Residual Plot with Anomalies
    threshold = 3 * residuals.std()
    mask = residuals.abs() > threshold
    num_anomalies = mask.sum()

    anomaly_df = pd.DataFrame({
        'Actual': y_test,
        'Predicted': y_pred,
        'Residual': residuals,
        'Anomaly': mask
    })

    plt.figure(figsize=(6, 4))
    plt.scatter(y_pred, residuals, label='Residuals', alpha=0.6)
    plt.scatter(y_pred[mask], residuals[mask], color='red', label='Anomalies')
    plt.axhline(0, color='red', linestyle='--')
    plt.title(f'{name} - Residual Plot with Anomalies\nâš  {num_anomalies} anomalies detected')
    plt.xlabel('Predicted')
    plt.ylabel('Residuals')
    plt.legend()
    plt.tight_layout()
    plt.show()

    print(f"ðŸ“‹ {name} - Anomaly Table:")
    print(anomaly_df[anomaly_df['Anomaly'] == True].sort_values(by='Residual', key=abs).round(2))

# MODEL COMPARISON TABLE
print("\nðŸ“Š Final Model Comparison Table:\n")
comparison_df = pd.DataFrame({
    'Model': list(results.keys()),
    'MSE': [results[m]['mse'] for m in results],
    'RÂ²': [results[m]['r2'] for m in results],
    'Accuracy (%)': [results[m]['accuracy'] for m in results]
}).sort_values(by='Accuracy (%)', ascending=False)

print(comparison_df.round(2))

# --------------------------------------------
# UPDATED Prophet Forecasting & Anomaly Detection (1 month ahead)
# --------------------------------------------
print(f"\n{'='*60}\nðŸ”® Starting Prophet Forecasting (1 Month)\n{'='*60}")

df_prophet = df[[datetime_col, target_col]].dropna().rename(columns={datetime_col: 'ds', target_col: 'y'})
df_prophet['ds'] = pd.to_datetime(df_prophet['ds'])

prophet_model = Prophet(
    changepoint_prior_scale=0.3,
    seasonality_mode='multiplicative',
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False
)
prophet_model.fit(df_prophet)

# Forecast 1 month ahead (720 hours)
future = prophet_model.make_future_dataframe(periods=720, freq='H')
forecast = prophet_model.predict(future)

# Filter only future forecast
last_date = df_prophet['ds'].max()
future_only = forecast[forecast['ds'] > last_date]

# Plot forecast
plt.figure(figsize=(12, 5))
plt.plot(future_only['ds'], future_only['yhat'], label='1-Month Forecast', color='green')
plt.fill_between(future_only['ds'], future_only['yhat_lower'], future_only['yhat_upper'], 
                 color='lightgreen', alpha=0.5, label='Confidence Interval')
plt.axvline(x=last_date, color='red', linestyle='--', label='Forecast Start')
plt.title('Prophet Forecast (Next 30 Days)')
plt.xlabel('Datetime'); plt.ylabel('Predicted Energy (kWh)')
plt.legend(); plt.tight_layout(); plt.show()

# Components
prophet_model.plot_components(forecast)
plt.show()

# Anomaly detection on training data only
historical = forecast[forecast['ds'] <= last_date]
merged = pd.merge(df_prophet, historical[['ds', 'yhat_lower', 'yhat_upper']], on='ds')
anomalies = merged[(merged['y'] < merged['yhat_lower']) | (merged['y'] > merged['yhat_upper'])]

print(f"\nâš  Prophet anomalies detected: {len(anomalies)}")
print(anomalies[['ds', 'y', 'yhat_lower', 'yhat_upper']])

plt.figure(figsize=(10, 5))
plt.plot(df_prophet['ds'], df_prophet['y'], label='Actual')
plt.plot(historical['ds'], historical['yhat_lower'], label='yhat_lower', linestyle='--', alpha=0.5)
plt.plot(historical['ds'], historical['yhat_upper'], label='yhat_upper', linestyle='--', alpha=0.5)
plt.scatter(anomalies['ds'], anomalies['y'], color='red', label='Anomalies')
plt.title('Prophet Historical Anomalies')
plt.xlabel('Datetime'); plt.ylabel('Energy (kWh)')
plt.legend()
plt.tight_layout()
plt.show()

# Save future forecast to CSV
future_only[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].to_csv('Energy_Forecast_1Month.csv', index=False)
print("ðŸ“ 1-Month Forecast saved to 'Energy_Forecast_1Month.csv'")
```
